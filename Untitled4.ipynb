```python
# Cell 2: Data & Features Engine
import pandas as pd
import numpy as np
import ta # Technical Analysis library
import requests
from datetime import datetime, timedelta
import py_vollib.black_scholes.greeks.analytical as greeks

class MarketDataEngine:
    def __init__(self):
        # 7% standard risk-free rate for Indian market
        self.risk_free_rate = 0.07

    def generate_dummy_data(self):
        """
        Jab tak aapke paas Upstox API keys na hon, tab tak test karne ke liye.
        Real data ke liye hum Upstox REST API use karenge.
        """
        print("Generating mock historical data for testing...")
        dates = pd.date_range(start='2024-01-01', periods=1000, freq='1min')
        np.random.seed(42)

        # Simulating random walk for NIFTY
        close = 21000 + np.random.randn(1000).cumsum() * 5

        df = pd.DataFrame({
            'datetime': dates,
            'open': close + np.random.randn(1000) * 2,
            'high': close + np.abs(np.random.randn(1000) * 5),
            'low': close - np.abs(np.random.randn(1000) * 5),
            'close': close,
            'volume': np.random.randint(1000, 10000, 1000),
            'iv': np.random.uniform(0.12, 0.20, 1000) # Implied Volatility between 12% to 20%
        })
        return df

    def add_technical_indicators(self, df):
        print("Adding Technical Features (MACD, RSI, VWAP)...")
        # MACD
        df['MACD'] = ta.trend.macd(df['close'])
        # RSI
        df['RSI'] = ta.momentum.rsi(df['close'], window=14)
        # VWAP
        df['VWAP'] = ta.volume.volume_weighted_average_price(
            high=df['high'], low=df['low'], close=df['close'], volume=df['volume']
        )
        # Returns for AI (Log returns are better for ML)
        df['log_return'] = np.log(df['close'] / df['close'].shift(1))

        return df

    def calculate_greeks(self, row, strike_price, option_type='c', days_to_expiry=7):
        """
        Black-Scholes formula se Option Greeks nikalna
        """
        S = row['close']
        K = strike_price
        t = days_to_expiry / 365.0
        r = self.risk_free_rate
        sigma = row['iv']

        try:
            delta = greeks.delta(option_type, S, K, t, r, sigma)
            theta = greeks.theta(option_type, S, K, t, r, sigma)
            return pd.Series({'Delta': delta, 'Theta': theta})
        except:
            return pd.Series({'Delta': 0.0, 'Theta': 0.0})

# --- Execution ---
engine = MarketDataEngine()

# 1. Get Data (Using mock data for now to ensure code runs without API keys)
df = engine.generate_dummy_data()

# 2. Add Tech Indicators
df = engine.add_technical_indicators(df)

# 3. Add Option Greeks (Assuming ATM strike price 21000 for Call Option)
print("Calculating Option Greeks...")
df[['Delta', 'Theta']] = df.apply(lambda row: engine.calculate_greeks(row, strike_price=21000, option_type='c'), axis=1)

# Drop rows with NaN values created by indicators (like first 14 days of RSI)
df = df.dropna().reset_index(drop=True)

print("\n✅ Feature Engineering Complete! AI ke liye Data tayar hai.")
display(df.tail()) # Shows the last 5 rows of the prepared data
```
